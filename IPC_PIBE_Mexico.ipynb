```python
import pandas as pd
import yfinance as yf
```

```python
ipc= ipc.drop(columns=["High","Low","Open","Volume"])
```

```python
#Rename
ipc=ipc.rename(columns={"Close":"Price"})
```

```python
ipc.head()
```

```python
#let's separate each president period
#Create a function
def separated_periods(date):
  if date >=pd.Timestamp ("1995-01-01") and date <=pd.Timestamp ("2000-12-31"):
    return "Ernesto Zedillo"
  elif date >=pd.Timestamp ("2001-01-01") and date <=pd.Timestamp ("2006-12-31"):
    return "Vicente Fox"
  elif date >=pd.Timestamp ("2007-01-01") and date <=pd.Timestamp ("2012-12-31"):
    return "Felipe Calderón"
  elif date >=pd.Timestamp ("2013-01-01") and date <=pd.Timestamp ("2018-12-31"):
    return "Peña Nieto"
  elif date >=pd.Timestamp ("2019-01-01") and date <=pd.Timestamp ("2024-12-31"):
    return "AMLO"
  elif date >=pd.Timestamp ("2025-01-01") and date <=pd.Timestamp ("2026-06-10"): #Current date
    return "Claudia Sheinbaum"
  else:
    return "0"
```

I needed to add pandas timestamp becasue python read the date's function as str

```python
#creating a new column
ipc["President"] =ipc["Date"].apply(separated_periods)
```

```python
def separated_periods_PIBE(Year):
  if Year >=1995 and Year <=2000:
    return "Ernesto Zedillo"
  elif Year >=2001 and Year <=2006:
    return "Vicente Fox"
  elif Year >=2007 and Year <=2012:
    return "Felipe Calderón"
  elif Year >=2013 and Year <=2018:
    return "Peña Nieto"
  elif Year >=2019 and Year <=2024:
    return "AMLO"
  elif  Year >=2025 and Year <=2026: #Current date
    return "Claudia Sheinbaum"
  else:
    return "No data"
```

```python
PIBE_per_state_long["Year"].dtype
```

```python
PIBE_per_state_long['Year'].unique()
```

```python
#We need to delete R and P from 2023 and 2024, then change the data tpye to int
PIBE_per_state_long["Year"] = (
    PIBE_per_state_long["Year"].astype(str)
    .str.replace(" R", "", regex=False)
    .str.replace(" P", "", regex=False))
#Convert to int
PIBE_per_state_long["Year"] = (
    pd.to_numeric(PIBE_per_state_long["Year"])
    .astype(int))
```

```python
print(PIBE_per_state_long["Year"].dtype)
print(PIBE_per_state_long["Year"].unique())
```

```python
PIBE_per_state_long["President"]=PIBE_per_state_long["Year"].apply(separated_periods_PIBE)
```

```python
PIBE_per_state_long.tail(50)
```

```python
#Remove blank spaces in price
PIBE_per_state_long = PIBE_per_state_long.dropna(subset=['Price'])
```

```python
PIBE_per_state_long['State'].unique()
```

```python
#Delete leading spaces
PIBE_per_state_long["State"] = (PIBE_per_state_long["State"].str.strip())
```

```python
#Drop values from state Concepto and  Estados Unidos Mexicanos
PIBE_per_state_long = PIBE_per_state_long[~PIBE_per_state_long['State'].isin(['Concepto', 'Estados Unidos Mexicanos'])]
```

```python
PIBE_per_state_long.to_csv('PIBE_per_state.csv',index=False)
```

```python
#i want to analyse the data with the growth per state for each presidental period
#I'm going to create a the grwoth and append it into new data
PIBE_per_state_long = PIBE_per_state_long.sort_values(
    ["State", "Year"])
```

```python
growth = (
    PIBE_per_state_long.groupby(['President','State'])
    .agg(Start = ('Price','first'),
    End=('Price','last')).reset_index()
    )
```

```python
#Calculate the growth
growth['GrowthPerState']=(
    (growth['End']- growth['Start'])/growth['Start']
) *100
```
